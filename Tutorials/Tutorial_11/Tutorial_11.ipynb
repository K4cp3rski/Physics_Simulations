```python
weights = np.zeros(shape=[points.shape[0], points.shape[0]])
distances = np.zeros(shape=[points.shape[0], points.shape[0]])
```

```python
for pos in positions:
    if adjacency[pos] == 0:
        continue
    else:
        dij = ((points[pos[1]][0] - points[pos[0]][0])**2 + (points[pos[1]][1] - points[pos[0]][1])**2)
        distances[pos] = dij
        weights[pos] = 1/dij
weights += weights.T
distances += distances.T
```

Here we'll imlement the algorithm for one ant

```python
def get_probas(pt):
    global alpha
    global beta
    adj = adjacency[:, pt]
    incident_pts = np.nonzero(adj)[0]
    inc_sum = 0
    probas = []
    for inc_pt in incident_pts:
        inc_sum += weights[pt, inc_pt]** alpha * pheromones[pt, inc_pt]**beta
    for inc_pt in incident_pts:
        pij = (weights[pt, inc_pt]** alpha * pheromones[pt, inc_pt]**beta)/(inc_sum)
        probas.append(pij)
    return incident_pts, probas
```

```python
def generate_path(i_node, dest_node):
    """
    i_node = beginng point
    dest_node = end point
    """
    path = []
    c_node = i_node
    path.append(c_node)
    # Current node
    while c_node != dest_node:
        incidence, probas = get_probas(c_node)
        n_node = np.random.choice(incidence, p=probas)
        path.append(n_node)
        c_node = n_node
    return path
```

```python
def rm_loop(path):
    while True:
        seen = set()
        uniq = [x for x in path if x not in seen and not seen.add(x)]
        if len(uniq) == len(path):
            break
        # print(uniq)
        dst = np.zeros(len(uniq))
        for num, el in enumerate(uniq):
            first_index = path.index(el)
            reversed_list = list(np.flip(path))
            first_index_in_reversed = reversed_list.index(el)
            last_index = len(path) -1 - first_index_in_reversed
            dst[num] = last_index - first_index
        # print(dst)
        # print(np.argmax(dst))
        el = np.argmax(dst)
        el = uniq[el]
        first_index = path.index(el)
        reversed_list = list(np.flip(path))
        first_index_in_reversed = reversed_list.index(el)
        last_index = len(path) -1 - first_index_in_reversed
        while last_index != first_index:
            path.pop(first_index)
            reversed_list = list(np.flip(path))
            first_index_in_reversed = reversed_list.index(el)
            last_index = len(path) -1 - first_index_in_reversed
    return path
```

```python
def calc_len(path):
    lp = 0
    for i in range(0, len(path)-1):
        lp += distances[path[i], path[i+1]]
    return lp
```

```python
def update_pheromones(paths_list, paths_len):
    global rho
    global pheromones
    for num, path in enumerate(paths_list):
        for i in range(0, len(path)-1):
            pos = (path[i], path[i+1])
            lgth = paths_len[num]
            pheromones[pos] += 1/lgth + (1 - rho)*(1/len(paths_list))
    pheromones += pheromones.T - np.diag(pheromones)
```
